```python
import requests
import pandas as pd
from io import StringIO
import datetime

end = datetime.date.today()
start = end - datetime.timedelta(700)
stockLabel = 'AAPL'
csv_url = 'https://query1.finance.yahoo.com/v7/finance/download/{stockLabel}?period1={start}&period2={end}&interval=1d&events=history'.format(
    stockLabel=stockLabel,
    start=start.strftime("%s"),
    end=end.strftime("%s")
)
print(csv_url)
req = requests.get(csv_url)
url_content = req.content
# print(url_content)
s = str(url_content, 'utf-8')

data = StringIO(s)

df = pd.read_csv(data).set_index('Date')

df
```
